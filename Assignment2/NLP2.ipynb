```python
!pip install gensim

import nltk
nltk.download('punkt')

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from gensim.models import Word2Vec

sentences = [
    "I love AI",
    "AI is future",
    "I love learning"
]



```

```python
# 1. Bag of Words

bow_vectorizer = CountVectorizer()
bow = bow_vectorizer.fit_transform(sentences)

print("Bag of Words Vocabulary:")
print(bow_vectorizer.get_feature_names_out())

print("\nBag of Words Matrix:")
print(bow.toarray())
```

```python
# 2. Normalized Count
import numpy as np

bow_array = bow.toarray()
normalized_bow = bow_array / bow_array.sum(axis=1, keepdims=True)

print("\nNormalized Bag of Words:")
print(normalized_bow)
```

```python
# 3. TF-IDF
tfidf_vectorizer = TfidfVectorizer()
tfidf = tfidf_vectorizer.fit_transform(sentences)

print("\nTF-IDF Matrix:")
print(tfidf.toarray())
```

```python
# 4. Word2Vec Embeddings
# Convert sentences into word lists
tokenized_sentences = [sentence.lower().split() for sentence in sentences]

model = Word2Vec(tokenized_sentences, vector_size=50, min_count=1)

print("\nWord2Vec Vector for 'ai':")
print(model.wv['ai'])
```
